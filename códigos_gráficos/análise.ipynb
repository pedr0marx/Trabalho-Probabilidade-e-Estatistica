```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("prec.xlsx")
```

```python
meses_colunas = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
df = df[~df[meses_colunas].eq('-').any(axis=1)] #Remove todas as linhas em que há valores nulos
```

```python
media_precipitacao = df[meses_colunas].mean()
```

```python
plt.figure(figsize=(10, 6))
media_precipitacao.plot(kind='bar', color='skyblue')
plt.title('Média de Precipitação Acumulada por Mês')
plt.xlabel('Mês')
plt.ylabel('Média de Precipitação Acumulada(mm)')
plt.xticks(rotation=45)  # Rotaciona os nomes dos meses no eixo x para melhor visualização
plt.grid(True)
plt.show()
```

```python
df_temp = pd.read_excel("temp.xlsx")
df_temp = df_temp[~df_temp[meses_colunas].eq('-').any(axis=1)]
```

```python
media_temperatura_media = df_temp[meses_colunas].mean()
```

```python
plt.figure(figsize=(10, 6))
media_temperatura_media.plot(kind='bar', color='red')
plt.title('Média das temperaturas médias compensadas por mês')
plt.xlabel('Mês')
plt.ylabel('Média de temperatura média compensada (°C)')
plt.xticks(rotation=45)  # Rotaciona os nomes dos meses no eixo x para melhor visualização
plt.grid(True)
plt.show()
```
